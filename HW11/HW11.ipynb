# Computing Lomb-Scargle periodograms for each time series

```
fourier_first = LombScargle(first['Julian Date'], first['Magnitude'], normalization='psd').autopower()
fourier_second = LombScargle(second['Julian Date'], second['Magnitude'], normalization='psd').autopower()
fourier_third = LombScargle(third['Julian Date'], third['Magnitude'], normalization='psd').autopower()
```

```
pl.plot(fourier_first[0], fourier_first[1], color = 'm')
pl.xscale('log')
pl.xlabel('Log of Wavelength')
pl.ylabel('Amplitude')
pl.title('Lomb-Scargle Periodogram for First Time Series')
pl.show()
```

**Figure 4**: Plot of the Lomb-Scargle Periodigram of the first time series comparing the log of the wavelength to the amplitude of the wavelength. 

```
pl.plot(fourier_second[0], fourier_second[1], color = 'm')
pl.xscale('log')
pl.xlabel('Log of Wavelength')
pl.ylabel('Amplitude')
pl.title('Lomb-Scargle Periodogram for Second Time Series')
pl.show()
```

**Figure 5**: Plot of the Lomb-Scargle Periodigram of the second time series comparing the log of the wavelength to the amplitude of the wavelength. 

```
pl.plot(fourier_third[0], fourier_third[1], color = 'm')
pl.xscale('log')
pl.xlabel('Log of Wavelength')
pl.ylabel('Amplitude')
pl.title('Lomb-Scargle Periodogram for Third Time Series')
pl.show()
```

**Figure 6**: Plot of the Lomb-Scargle Periodigram of the third time series comparing the log of the wavelength to the amplitude of the wavelength.

# Phase Folding

```
pfold_first = pd.DataFrame()
```

```
pfold_first['phase'] = np.mod(first['Julian Date'], 1/fourier_first[0][70])
pfold_first['dmag'] = first['Magnitude']
```

```
pl.figure(figsize=(10,10))
pl.plot(pfold_first.phase, pfold_first.dmag, ',', color = 'm')
pl.title('Phase Folding of First Time Series', size = 20)
pl.ylabel('DMag', size = 20)
pl.xlabel('Phase', size = 20)
```

**Figure 10**: Plot of the phase folding for the first time series comparing the magnitude of the time series to the phase of the time series. 

```
pfold_second = pd.DataFrame()
pfold_second['phase'] = np.mod(second['Julian Date'], 1/fourier_second[0][72])
pfold_second['dmag'] = second['Magnitude']
pl.figure(figsize=(10,10))
pl.plot(pfold_second.phase, pfold_second.dmag, ',', color = 'm')

pl.title('Phase Folding of Second Time Series', size = 20)
pl.ylabel('DMag', size = 20)
pl.xlabel('Phase', size = 20)
```

**Figure 11**: Plot of the phase folding for the second time series comparing the magnitude of the time series to the phase of the time series. 

```
pfold_third = pd.DataFrame()
pfold_third['phase'] = np.mod(third['Julian Date'], 1/fourier_third[0][108])
pfold_third['dmag'] = third['Magnitude']
pl.figure(figsize=(10,10))
pl.plot(pfold_third.phase, pfold_third.dmag, ',', color = 'm')

pl.title('Phase Folding of Third Time Series', size = 20)
pl.ylabel('DMag', size = 20)
pl.xlabel('Phase', size = 20)
```

**Figure 12**: Plot of the phase folding for the second time series comparing the magnitude of the time series to the phase of the time series. 

# Binning

```
means_first, edges_first, binnum_first = stats.binned_statistic(pfold_first.phase, pfold_first.dmag, bins=20)

means_second, edges_second, binnum_second = stats.binned_statistic(pfold_second.phase, pfold_second.dmag, bins=20)

means_third, edges_third, binnum_third = stats.binned_statistic(pfold_third.phase, pfold_third.dmag, bins=20)
```

```
pl.plot(pfold_first.phase, pfold_first.dmag, ',', zorder=0, color = 'm')
pl.hlines(means_first, edges_first[:-1], edges_first[1:], 'k', lw=4, zorder=1)
pl.title('Binning of Phase Folding for First Time Series')
pl.xlabel('Phase')
pl.ylabel('Dmag')

pl.show()
```

**Figure 13**: Plot of the binning of the phase folding for the first time series comparing the magnitude of the time series to the phase of the time series.

```
pl.plot(pfold_second.phase, pfold_second.dmag, ',', zorder=0, color = 'm')
pl.hlines(means_second, edges_second[:-1], edges_second[1:], 'k', lw=4, zorder=1)
pl.title('Binning of Phase Folding for Second Time Series')
pl.xlabel('Phase')
pl.ylabel('Dmag')

pl.show()
```

**Figure 14**: Plot of the binning of the phase folding for the second time series comparing the magnitude of the time series to the phase of the time series.

```
pl.plot(pfold_third.phase, pfold_third.dmag, ',', zorder=0, color = 'm')
pl.hlines(means_third, edges_third[:-1], edges_third[1:], 'k', lw=4, zorder=1)
pl.title('Binning of Phase Folding for Third Time Series')
pl.xlabel('Phase')
pl.ylabel('Dmag')

pl.show()
```

**Figure 15**: Plot of the binning of the phase folding for the third time series comparing the magnitude of the time series to the phase of the time series.
